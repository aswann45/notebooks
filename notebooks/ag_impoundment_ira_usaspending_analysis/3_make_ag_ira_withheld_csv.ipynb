```python
from pathlib import Path

import pandas as pd
```

```python
pd.options.display.float_format = "{:,.0f}".format
```

```python
data_dir: Path = Path().home() / "data" / "usaspending"
```

```python
df = pd.read_csv(data_dir / "ag_ira_awards.csv")
```

```python
df["max_promised_amount"] = df[
    ["total_obligated_amount", "total_outlayed_amount_for_overall_award"]
].max(axis=1)
df["max_promised_amount"]
```

```python
df.loc[df["treasury_accounts_funding_this_award"].str.contains("1908")][
    "treasury_accounts_funding_this_award"
].value_counts()
```

```python
df["total_outlayed_amount_for_overall_award"] = df[
    "total_outlayed_amount_for_overall_award"
].fillna(0)
```

```python
df["max_withheld_amount"] = (
    df["max_promised_amount"] - df["total_outlayed_amount_for_overall_award"]
)
df
```

```python
df.loc[df["treasury_accounts_funding_this_award"].str.contains("1908")][
    "treasury_accounts_funding_this_award"
].value_counts()
```

```python
df.loc[df["treasury_accounts_funding_this_award"].str.contains("1908")][
    ["treasury_accounts_funding_this_award", "max_withheld_amount"]
].value_counts()
```

```python
df.loc[
    df["treasury_accounts_funding_this_award"].str.contains("1908")
    & (df["max_withheld_amount"] != 0)
][["treasury_accounts_funding_this_award", "max_withheld_amount"]]
```

```python
df_withheld = df.loc[df.max_withheld_amount > 0]
```

```python
df_withheld.loc[
    df_withheld["treasury_accounts_funding_this_award"].str.contains("1908")
]["treasury_accounts_funding_this_award"].value_counts()
```

```python
df_withheld.to_csv(data_dir / "ag_ira_awards_withheld.csv", index=False)

df_withheld
```
