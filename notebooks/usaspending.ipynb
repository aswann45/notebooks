```python
from pathlib import Path

import pandas as pd
```

```python
data_dir: Path = Path().home() / "data" / "usaspending"

subdirs: list[Path] = [
    item for item in data_dir.iterdir() if item.is_dir()
]
```

```python
dfs: list[pd.DataFrame] = []
for i in data_files:
    df = pd.read_csv(i, usecols=cols)
    dfs.append(df)
df_out = pd.concat(dfs)
df_out.to_csv(data_dir / "ag_awards.csv", index=False)

del dfs
del df_out

df = pd.read_csv(data_dir / "ag_awards.csv")
```

* action_date_fiscal_year: fy in which the most recent change to the award occured; should be 2022 or later
* treasury_accounts_funding_this_award: tafs associated with the award; should be one of the IRA/IIJA tafs

```python
df = df.sort_values(["total_outlayed_amount_for_overall_award", "total_obligated_amount"], ascending=False).drop_duplicates("assistance_award_unique_key")
df
```

```python
df = df.loc[df["action_date_fiscal_year"] >= 2022]
df
```

```python
df = df.drop(index=df.loc[df["total_obligated_amount"] < 100].index)
df
```

```python
df = df.dropna(subset=["treasury_accounts_funding_this_award"])
df
```

```python
tafs = [
    # osec
    "012-2022/2031-0115",
    "012-2024/2028-0115",
    # fsa
    "012-2022/2031-1124",
    "012-2022/2031-1140",
    # nifa
    "012-2022/2031-1500",
    # nrcs
    "012-2022/2031-1000",
    "012-2023/2031-1004",
    "012-2024/2031-1004",
    "012-2025/2031-1004",
    "012-2026/2031-1004",
    # rcbs
    "012-2022/2031-2073",
    "012-2022/2031-1908",
    "012-2023/2031-1908",
    "012-2024/2031-1908",
    "012-2025/2031-1908",
    "012-2026/2031-1908",
    # rus
    "012-2023/2031-1230",
]
```

```python
mask = df["treasury_accounts_funding_this_award"].str.contains("|".join(tafs))
df_ira = df.loc[mask]
df_ira
```

```python
df_ira["max_promised_amount"] = df_ira[["total_obligated_amount", "total_outlayed_amount_for_overall_award"]].max(axis=1)
df_ira
```

```python
df_ira["max_withheld_amount"] = df_ira["max_promised_amount"] - df_ira["total_outlayed_amount_for_overall_award"]
df_ira
```

```python
df_withheld = df_ira.loc[df_ira.max_withheld_amount > 0]

df_withheld.to_csv(data_dir / "ag_awards_fy2022_fy2025_withheld.csv", index=False)

df_withheld
```
